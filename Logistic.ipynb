## Start with data preprocessing


```python
df = pd.read_csv('data2.csv')
print(df.head())
```

```python
null = df.isna().mean() * 100
print("Percentage of missing values in each column:\n", null)
```

Cloud9AM, Cloud3PM, Evaporatoin and Sunshine has really high null value rate, so imputation isn't very practical here. SO we either need to drop the columns or perform feature engineering.

First I am gonna try to combine the Cloud

```python
df["CloudMean"] = df[["Cloud9am", "Cloud3pm"]].mean(axis=1).round()
df["CloudMean"] = df["CloudMean"].fillna(df["CloudMean"].mode()[0])

print(df["CloudMean"].isna().mean() * 100)

for col in ["CloudMean"]:
    col_count = df[col].value_counts().sort_index()
    print(col_count)

df = df.drop(columns=["Cloud9am", "Cloud3pm"])
df.info
```

Replacing still missing values with mode as it is categorical type data 

```python
df["CloudMean"] = df["CloudMean"].fillna(df["CloudMean"].mode()[0])
print(df["CloudMean"].isna().mean() * 100)

for col in ["CloudMean"]:
    col_count = df[col].value_counts().sort_index()
    print(col_count)
```
